<h2> Train Test Split </h2>

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2> Linear Regression </h2>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
X_train
```

```python
y_train
```

```python
sns.regplot(x= y_test, y= y_pred, line_kws={"color": "red"})
plt.ylabel('Predicted Price')
plt.xlabel('Actual Price')
```

```python
y_train_pred = lr.predict(X_train)
```

```python
sns.regplot(x= y_train, y= y_train_pred, line_kws={"color": "red"})
plt.ylabel('Predicted Price')
plt.xlabel('Actual Price')
```

<h2> Model Evluation </h2>

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error

print(f" The Mean_absolute_error for this model is {mean_absolute_error(y_pred,y_test)}")
print(f" The Mean Sqaured Error is {mean_squared_error(y_pred, y_test)}")
print(f" The r2_score_error for this model is {r2_score(y_pred,y_test)}")
```
